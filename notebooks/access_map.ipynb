```python
import pandas as pd
df = pd.read_csv('neuronbridge_accesses.csv')
max_count = df['Count'].max()
df.head()
```

```python
import branca.colormap as cmp
colorscale = cmp.LinearColormap(
    ['blue', 'red'],
    vmin=1, vmax=1000,
    caption='Color Scale for Map' #Caption for Color scale or Legend
)
colorscale
```

```python
import branca
import folium
import math
from folium.plugins import MarkerCluster

#colorscale = branca.colormap.linear.YlOrRd_09.scale(0, df['Count'].max())

f = folium.Figure(width=1100, height=800)
world_map = folium.Map(tiles="cartodbpositron", max_bounds=True).add_to(f)
#marker_cluster = MarkerCluster().add_to(world_map)

def style_function(feature):
    return {
        'fillOpacity': 0.5,
    }

for i in range(len(df)):
    lat = df.iloc[i]['Latitude']
    long = df.iloc[i]['Longitude']
    count = df.iloc[i]['Count']
    #color = "#aa00aa"
    color = colorscale(count)
    radius = 7
    #radius = (max_count/count) * count
    folium.CircleMarker(location = [lat, long], color=color, radius=radius, tooltip=f"{count}", fill=True, opacity=0.001).add_to(world_map)#.add_to(marker_cluster)

colorscale.caption = "Number of API accesses per location"
colorscale.add_to(world_map)
world_map
```
